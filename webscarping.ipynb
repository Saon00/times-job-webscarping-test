```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
# create csv file
outFile = open('timesJobs_scrap.csv', 'w', encoding='utf8', newline='')
writer = csv.writer(outFile)
writer.writerow(['Company Name','Job Name','Job Skills','Experience--Salary--Location','Details Link'])
```

```python
# setting the link
url = 'https://www.timesjobs.com/jobskill/java-jobs'
r = requests.get(url)
soup = BeautifulSoup(r.content, 'html.parser')
```

```python
jobs = soup.find_all('li', class_='clearfix joblistli')

for index, job in enumerate(jobs, 1):
    job_name = job.find('a')
    jn = job_name.text.replace('\t', '').replace('\r\n','')

    company_name = job.find('h3', class_='joblist-comp-name')
    cn = company_name.text.replace('\t', '').replace('\r\n','').replace('(More Jobs)\n', '')

    job_skills = job.find('span', class_='srp-skills')
    js = job_skills.text.replace('\n','').replace(' ', ', ').strip()

    more_infos = job.find('ul', class_='job-more-dtl clearfix')
    mi = more_infos.text.replace('\n',' ').replace('\r\n','').replace('\t','').replace('\r','').strip()

    job_details = job.header.h2.a['href']
    jd = job_details

    print(f'''{index}.
    Company Name: {cn}
    Job Name: {jn}
    Job Skills: {js}
    Experience--Salary--Location: {mi}
    Details Link: {jd}
    ''')
    
    writer.writerow([cn, jn, js, mi, jd])

outFile.close()
```
